```python
import matplotlib.pyplot as plt

# Data
# name of slice (categories)
labels = ['Apple', 'Banana', 'Cherry', 'Mango']
 # Ye values hain jo har fruit ka proportion batati hain
sizes = [30, 25, 20, 25]
#color of slice
colors = ['red', 'yellow', 'pink', 'orange']
#for highlight the apple slice
explode = (0.1, 0, 0, 0)

# Pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)

# Title
plt.title("Fruit Distribution")

# Equal aspect ratio ensures pie is drawn as a circle.
plt.axis('equal')

# Show chart
plt.show()
```
